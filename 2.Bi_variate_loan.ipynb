```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("loan_approval.csv")
```

```python
dataset
```

```python
def quanQual(dataset):
    qual=[]
    quan=[]
    for colName in dataset.columns:
        if(dataset[colName].dtype=='O'):
            qual.append(colName)
        else:
            quan.append(colName)
    return qual,quan
    
```

```python
qual,quan=quanQual(dataset)
```

```python
quan
```

```python
dataset.describe()
```

```python
descriptive=pd.DataFrame(index=["mean","median","mode","Q1:25%","Q2:50%","Q3:75%","Q4:100%","IQR",
                                "1.5 rule","Lesser","Greater","min","max","kurtosis",
                               "skew","var","Std"],columns=quan)
for col_Name in quan:
    descriptive[col_Name]["mean"]=dataset[col_Name].mean()
    descriptive[col_Name]["median"]=dataset[col_Name].median()
    descriptive[col_Name]["mode"]=dataset[col_Name].mode()[0]
    descriptive[col_Name]["Q1:25%"]=dataset.describe()[col_Name]["25%"] # column and row name
    descriptive[col_Name]["Q2:50%"]=dataset.describe()[col_Name]["50%"]
    descriptive[col_Name]["Q3:75%"]=dataset.describe()[col_Name]["75%"]
    descriptive[col_Name]["Q4:100%"]=dataset.describe()[col_Name]["max"]
    descriptive[col_Name]["IQR"]=descriptive[col_Name]["Q3:75%"]-descriptive[col_Name]["Q1:25%"]
    descriptive[col_Name]["1.5 rule"]=1.5*descriptive[col_Name]["IQR"]
    descriptive[col_Name]["Lesser"]=descriptive[col_Name]["Q1:25%"]- descriptive[col_Name]["1.5 rule"]
    descriptive[col_Name]["Greater"]=descriptive[col_Name]["Q3:75%"]+descriptive[col_Name]["1.5 rule"]
    descriptive[col_Name]["min"]=dataset[col_Name].min()
    descriptive[col_Name]["max"]=dataset[col_Name].max()
    descriptive[col_Name]["kurtosis"]=dataset[col_Name].kurtosis()
    descriptive[col_Name]["skew"]=dataset[col_Name].skew()
    descriptive[col_Name]["var"]=dataset[col_Name].var()
    descriptive[col_Name]["Std"]=dataset[col_Name].std()
```

```python
descriptive
```

```python
quan
```

```python
lesser_outlier_true=[]
lesser_outlier_false=[]
for col_Name in quan:
    if(descriptive[col_Name]["min"]<descriptive[col_Name]["Lesser"]):
        lesser_outlier_true.append(col_Name)
    else:
        lesser_outlier_false.append(col_Name)
        
```

```python
lesser_outlier_true
```

```python
lesser_outlier_false
```

```python
greater_outlier_true=[]
greater_outlier_false=[]
for col_Name in quan:
    if(descriptive[col_Name]["max"]>descriptive[col_Name]["Greater"]):
        greater_outlier_true.append(col_Name)
    else:
        greater_outlier_false.append(col_Name)
```

# Frequency table

```python
freq_table=pd.DataFrame(columns=["Unique_values","Frequency","Relative_frequency","Cumulative_frequency"])
```

```python
freq_table
```

```python
freq_table["Unique_values"]=dataset[ ' cibil_score'].value_counts().index
freq_table["Frequency"]=dataset[' cibil_score'].value_counts().values
```

```python
freq_table
```

```python
freq_table["Relative_frequency"]=freq_table["Frequency"]/601
freq_table["Cumulative_frequency"]=freq_table["Relative_frequency"].cumsum()
```
