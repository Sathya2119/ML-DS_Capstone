```python
def quanQual(dataset):
    qual=[]
    quan=[]
    for colName in dataset.columns:
        if(dataset[colName].dtype=='O'):
            qual.append(colName)
        else:
            quan.append(colName)
    return quan,qual
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calcu_vif(X):
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(X.shape[1])]
    return vif
```

```python
calcu_vif(dataset[['loan_id',' no_of_dependents',' income_annum',' loan_amount',
 ' loan_term',' cibil_score',' residential_assets_value',' commercial_assets_value',
 ' luxury_assets_value',' bank_asset_value']])
```

# T-test (independent-sample - unpaired T-test)
different group (Self_Employed-->( yes  ) but same condition(Loan_amount)

```python
from scipy.stats import ttest_ind
self_emp_yes=dataset[dataset[' self_employed']=='Yes'][ ' luxury_assets_value']
self_emp_no=dataset[dataset[' self_employed']=='No'][' luxury_assets_value']
ttest_ind(self_emp_yes,self_emp_no)
```

```python
from scipy.stats import ttest_rel
self_emp1=dataset[dataset[' self_employed']=='Yes'][' income_annum']
self_emp2=dataset[dataset[' self_employed']=='Yes'][' luxury_assets_value']
ttest_rel(self_emp1,self_emp2)
```
